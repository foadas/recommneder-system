```python
import hazm
import pandas as pd
import tensorflow as tf

df = pd.read_csv('/content/drive/MyDrive/data/2-comments.csv', encoding='utf-8', usecols=['product_id', 'user_id', 'product_title', 'recommend'])

df = df.fillna({'recommend': '', 'product_title': ''})

df['recommend'] = df['recommend'].replace('\\N', 'no_idea')

recommend_mapping = {
    'not_recommended': 0,
    'no_idea': 3,
    'recommended': 5
}

df['recommend'] = df['recommend'].map(recommend_mapping)

normalizer = hazm.Normalizer()
stop_words = { 'سری', "به", "و", "ای", "عددی", "مدل", "کد" }

def hazm_custom_standardization(text):

    text = normalizer.normalize(text)

    tokens = hazm.word_tokenize(text)

    filtered_tokens = [token for token in tokens if token not in stop_words]

    return ' '.join(filtered_tokens)

df['product_title'] = df['product_title'].apply(hazm_custom_standardization)


ratings_dataset = tf.data.Dataset.from_tensor_slices(dict(df))


assert isinstance(ratings_dataset, tf.data.Dataset)

# Check the dataset size
print(f"Dataset size: {len(df)}")

# Display some samples
head = ratings_dataset.take(5)
for rate in head.as_numpy_iterator():
    print(rate)
```

```python
# Split dataset randomly (80% for train and 20% for test)
trainset_size = 0.8 * ratings_dataset.__len__().numpy()

tf.random.set_seed(42)

ratings_dataset_shuffled = ratings_dataset.shuffle(

    buffer_size=100_000,

    seed=42,

    reshuffle_each_iteration=False
)

ratings_trainset = ratings_dataset_shuffled.take(trainset_size)
ratings_testset = ratings_dataset_shuffled.skip(trainset_size)

print(
    "ratings_trainset size: %d" % ratings_trainset.__len__()
)
print(
    "ratings_testset size: %d" % ratings_testset.__len__()
)
```

```python
recommend_normalization_layer = tf.keras.layers.Normalization(axis=None)

recommend_normalization_layer.adapt(
    ratings_trainset.map(
        lambda x: x['recommend']
    )
)
```

```python
user_id_lookup_layer = tf.keras.layers.IntegerLookup(mask_token=None)


user_id_lookup_layer.adapt(
    ratings_trainset.map(
        lambda x: x['user_id']
    )
)

print(
    f"Vocabulary[:10] -> {user_id_lookup_layer.get_vocabulary()[:10]}"
)
```

```python
user_id_model = tf.keras.Sequential(
    [
        user_id_lookup_layer,
        user_id_embedding_layer
    ]
)

user_ids = [764992, 1152044, 655]
user_ids_tensor = tf.convert_to_tensor(user_ids, dtype=tf.int32)


embeddings = user_id_model(user_ids_tensor)


print("Embeddings for user ids:", user_ids)
print(embeddings.numpy())
```

```python
product_id_lookup_layer = tf.keras.layers.IntegerLookup(mask_token=None)
product_id_lookup_layer.adapt(
    ratings_trainset.map(
        lambda x: x['product_id']
    )
)

# Same as user_id_embedding_dim to be able to measure the similarity
product_id_embedding_dim = 32

product_id_embedding_layer = tf.keras.layers.Embedding(
    input_dim=product_id_lookup_layer.vocabulary_size(),
    output_dim=product_id_embedding_dim
)

product_id_model = tf.keras.Sequential(
    [
        product_id_lookup_layer,
        product_id_embedding_layer
    ]
)
```

```python
import re

# Function to remove specific words and numbers from product titles
def custom_standardization(input_text):
    # Define the words you want to remove
    words_to_remove = [ 'سری',"به","و","ای","عددی","مدل", "کد"]

    # Remove the specific words
    for word in words_to_remove:
        input_text = tf.strings.regex_replace(input_text, word, "")

    # Remove any numeric values
    input_text = tf.strings.regex_replace(input_text, r'\d+', "")

    # Optional: Remove extra spaces
    input_text = tf.strings.regex_replace(input_text, r'\s+', " ")



    # Convert to lowercase (if needed)
    input_text = tf.strings.strip(input_text)

    return input_text

out = custom_standardization('لپ تاپ ایران')
print (out)
```

```python


product_title_vectorization_layer = \
    tf.keras.layers.TextVectorization()
product_title_vectorization_layer.adapt(
    ratings_trainset.map(
        lambda rating: rating['product_title']
    )
)

print(
    "Vocabulary[40:50] -> ",
    product_title_vectorization_layer.get_vocabulary()[40:50]
)


product_title_embedding_dim = 32
product_title_embedding_layer = tf.keras.layers.Embedding(
    input_dim=len(product_title_vectorization_layer.get_vocabulary()),
    output_dim=product_title_embedding_dim,
    # Whether or not the input value 0 is a MASK token.
    mask_zero=True
)

product_title_model = tf.keras.Sequential(
    [
       product_title_vectorization_layer,
       product_title_embedding_layer,
       tf.keras.layers.GlobalAveragePooling1D()
    ]
)
```

```python
import tensorflow_recommenders as tfrs

retrieval_task_layer = tfrs.tasks.Retrieval(
    metrics = tfrs.metrics.FactorizedTopK(

    corpus.batch(128).map(candidate_model)
)
)
```

```python
class RetrievalModel(tfrs.models.Model):

  def __init__(self, query_model, candidate_model, retrieval_task_layer):
    super().__init__()
    self.query_model: tf.keras.Model = query_model
    self.candidate_model: tf.keras.Model = candidate_model
    self.retrieval_task_layer: tf.keras.layers.Layer = retrieval_task_layer

  def compute_loss(self, features, training=False) -> tf.Tensor:
    query_embeddings = self.query_model(features['user_id'])
    positive_candidate_embeddings = self.candidate_model(
        {"product_id": features["product_id"], "product_title": features["product_title"]}
    )

    loss = self.retrieval_task_layer(
        query_embeddings,
        positive_candidate_embeddings
    )
    return loss
```

```python
digikala_retrieval_model = RetrievalModel(
    query_model,
    candidate_model,
    retrieval_task_layer,
    #title_model,
)

optimizer_step_size = 0.1
digikala_retrieval_model.compile(
    optimizer=tf.keras.optimizers.Adagrad(
        learning_rate=optimizer_step_size
    )
)
```

```python
# Query tower
query_model1 = user_id_model

# Candidate tower
candidate_model1 = product_id_model

product_title_model1 = product_title_model
```

```python
class RankingModel(tfrs.models.Model):

  def __init__(self, query_model, candidate_model):
    super().__init__()

    self.query_model: tf.keras.Model = query_model1
    self.candidate_model: tf.keras.Model = candidate_model1
    self.product_model: tf.keras.Model = product_title_model1
    self.rating_model = tf.keras.Sequential(
        [
            tf.keras.layers.Dense(256, activation='relu'),
            tf.keras.layers.Dense(64, activation='relu'),
            tf.keras.layers.Dense(1)
        ]
    )
    self.ranking_task_layer: tf.keras.layers.Layer = tfrs.tasks.Ranking(
        loss=tf.keras.losses.MeanSquaredError(),
        metrics=[
            tf.keras.metrics.RootMeanSquaredError()
        ]
    )


  def compute_loss(self, features, training=False) -> tf.Tensor:
    query_embeddings = self.query_model(features['user_id'])
    candidate_embeddings = self.candidate_model(features["product_id"])
    product_title_embeddings = self.product_model(features['product_title'])
    rating_predictions = self.rating_model(
        tf.concat(
            [query_embeddings, candidate_embeddings, product_title_embeddings],
            axis=1
        )
    )

    loss = self.ranking_task_layer(
        predictions=rating_predictions,
        labels=features["recommend"]
    )
    return loss

  def call(self, inputs):
        user_embeddings = self.query_model(inputs['user_id'])
        product_embeddings = self.candidate_model(inputs['product_id'])
        title_embeddings = self.product_model(inputs['product_title'])

        combined_embeddings = tf.concat([user_embeddings, product_embeddings, title_embeddings], axis=1)

        rating_predictions = self.rating_model(combined_embeddings)

        return rating_predictions
```

```python
digikala_ranking_model = RankingModel(query_model, candidate_model)

optimizer_step_size = 0.1
digikala_ranking_model.compile(
    optimizer=tf.keras.optimizers.Adagrad(
        learning_rate=optimizer_step_size
    )
)
```

```python
ranking_ratings_trainset = ratings_trainset.shuffle(100_000).batch(8192).cache()
ranking_ratings_testset = ratings_testset.batch(4096).cache()

history = digikala_ranking_model.fit(
    ranking_ratings_trainset,
    validation_data=ranking_ratings_testset,
    validation_freq=1,
    epochs=5
)
```

```python
import numpy as np

all_user_ids = []
def print_all_user_ids_flat(dataset):


    for batch in dataset:

        user_ids = batch["user_id"].numpy()
        all_user_ids.extend(user_ids)


    print(np.array(all_user_ids))


print_all_user_ids_flat(retrieval_testset)
```

```python
def calculate_recall(true_positives):


    # Calculate False Negatives
    false_negatives = 10 - true_positives

    # Calculate Recall
    recall = true_positives / (true_positives + false_negatives)



    return recall

true_positives = 8

recall = calculate_recall(true_positives)

print(f"Recall: {recall:.2f}")
```

```python
import tensorflow as tf
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity


for user_id in all_user_ids:
    print(user_id)
    user_id_tensor = user_id

    afinity_scores, candidate_product_ids = brute_force_layer(
        tf.constant([user_id])
    )

    candidate_product_ids = candidate_product_ids[0, :10]

    item_ids = candidate_product_ids


    item_ids_tensor = tf.convert_to_tensor(item_ids, dtype=tf.int64)


    def get_item_embeddings(item_ids_tensor):

        return tf.random.uniform((len(item_ids_tensor), 10))


    item_embeddings = get_item_embeddings(item_ids_tensor)


    item_embeddings_np = item_embeddings.numpy()


    cosine_sim_matrix = cosine_similarity(item_embeddings_np)


    threshold = 0.6


    num_items = len(item_ids)
    similar_item_counts = np.zeros(num_items)

    for i in range(num_items):

        similar_items = np.sum(cosine_sim_matrix[i] >= threshold) - 1
        similar_item_counts[i] = similar_items

    # Aggregate results
    total_similar_items = np.sum(similar_item_counts)
    average_similar_items = np.mean(similar_item_counts)

    print(f"User ID: {user_id_tensor}")  # Use 'user_id_tensor'
    print(f"Total number of similar items across all items: {total_similar_items}")
    print(f"Average number of similar items per item: {average_similar_items}")

    aggregated_result = total_similar_items
    print(f"Aggregated result (total similar items): {aggregated_result}")


    recall = calculate_recall(average_similar_items)

    print(f"Recall: {recall:.2f}")
```
